```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib as mpl
np.random.seed(77)
```

```python
def fgm(l0, L, b, C, x, U, N=100):
    q = 2
    m, n = C.shape
    l = l0.copy()
    z = l0.copy()
    x_hat = np.zeros(n)
    
    U_val = []
    constr_val = []
    
    for t in range(N):
        y = np.clip(l - (1/L) * (b - C @ x(l)), 0, None)
        z -= (1/L) * ((t + 1)/2) * (b - C @ x(l))
        x_hat += ((t + 1)/2) * x(l)
        l = (2/(t + 3)) * np.clip(z, 0, None) + (1 - 2/(t + 3)) * y
        
        U_val.append(U(x_hat * (4/((t + 1) * (t + 2))) ))
        constr_val.append(np.linalg.norm(np.clip(C @ x_hat * (4/((t + 1) * (t + 2))) - b, 0, None), q))
        
    return U_val, constr_val
```

```python
def mirror(x0, M, b, C, grad_u, eps, U, N=100):
    p, q = 2, 2
    m, n = C.shape
    L = max(np.linalg.norm(C[j], p)**2 for j in range(m))
    x = x0.copy()
    x_hat = x0.copy()
    I, J = 0, 0
    
    U_val = []
    constr_val = []
    
    for t in range(N):
        if np.all(C @ x - b <= eps):
            i = random.randint(0, n-1)
            x[i] -= min(0, (eps / M**2) * grad_u[i](x[i]))
            x_hat += x
            I += 1
            U_val.append(U((1/I) * x_hat))
            constr_val.append(np.linalg.norm(np.clip(C @ ((1/I) * x_hat) - b, 0, None), q))
        else:
            j = np.argmax(C @ x - b)
            i = np.random.choice(np.nonzero(C[j])[0])
            x[i] -= np.clip((eps / L**2) * C[j][i], 0, None)
            J += 1
            U_val.append(U(x))
            constr_val.append(np.linalg.norm(np.clip(C @ x - b, 0, None), q))
        
    return U_val, constr_val
```

```python
m = 40
n = 100
```

```python
l0 = np.random.random(m)
```

```python
C = np.random.choice(2, size=(m, n), p=[0.999, 0.001])
b = np.round(np.random.random(m) * n)
```

```python
a = np.random.randint(1, 50, size=n)
sigma = 0.001
```

```python
u = [lambda x: a[k] * x - (sigma * n / 2) * x**2 for k in range(n)]
grad_u = [lambda x: a[k] - (sigma * n) * x for k in range(n)]
```

```python
U = lambda x: sum(u[i](x[i]) for i in range(n))
```

```python
x = lambda l: np.clip(a - C.T @ l, 0, None) / (n * sigma)
```

```python
L = np.nonzero(C)[0].shape[0] / (n * sigma)
```

```python
L
```

```python
U_val_fgm, constr_val_fgm = fgm(l0.copy(), L, b, C, x, U, N=150)
```

```python
U_val_mirror, constr_val_mirror = mirror(x(l0), L, b, C, grad_u, 0.2, U, N=15_000)
```

```python
itm = len(U_val_mirror)
it = len(U_val_fgm)
```

```python
print('SMD:', U_val_mirror[-1], constr_val_mirror[-1])
print('FGM:', U_val_fgm[-1], constr_val_fgm[-1])
```

```python
mpl.style.use("default")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))

ax1.plot(np.linspace(0, itm-1, itm), U_val_mirror, 'C2--', linewidth=1)
ax1.plot(np.linspace(0, it-1, it) * n, U_val_fgm, 'C3-.', linewidth=1)
ax1.set_ylim(U_val_fgm[-1] * 0.99, U_val_mirror[0] * 1.005)

ax1.grid(alpha=0.35, linewidth=1)
ax1.set_xlabel('Oracle calls')
ax1.set_ylabel('$U(x)$')

ax2.plot(np.linspace(0, itm-1, itm), constr_val_mirror, 'C2--', label='Fast Gradient Method', linewidth=1)
ax2.plot(np.linspace(0, it-1, it) * n, constr_val_fgm, 'C3-.', label='Stochastic Mirror Descent', linewidth=1)
ax2.set_ylim(-20., constr_val_fgm[0] * 1.07)

ax2.legend()
ax2.grid(alpha=0.35, linewidth=1)
ax2.set_xlabel('Oracle calls')
ax2.set_ylabel('$||(C x - b)_+||_2$')

fig.text(0.3, 0.85, 'FGM: U(x) = 512\'458\n         g(x) = 0.58')
fig.text(0.3, 0.75, 'SMD: U(x) = 512\'467\n         g(x) = 0.42')

fig.tight_layout()
```
